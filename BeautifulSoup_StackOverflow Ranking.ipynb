```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
url="https://stackexchange.com/leagues/1/alltime/stackoverflow/2008-07-31?sort=reputationchange&page=1"
```

```python
soup=bs(requests.get(url).content,"html.parser")
```

```python
df=pd.DataFrame()
for i in range(1,10):
    url="https://stackexchange.com/leagues/1/alltime/stackoverflow/2008-07-31?sort=reputationchange&page="+str(i)
    soup=bs(requests.get(url).content,"html.parser")
    data=[]
    ranks=soup.select(".contentWrapper div.league-container div:nth-child(1) span.number")
    names=soup.select(".contentWrapper div.league-container div.userInfo h2 a")
    golds=soup.select(".contentWrapper div.league-container div.userInfo span:nth-child(2) span.badgecount")
    silvers=soup.select(".contentWrapper div.league-container div.userInfo span:nth-child(3) span.badgecount")
    bronzes=soup.select(".contentWrapper div.league-container div.userInfo span:nth-child(4) span.badgecount")
    memberships=soup.select(".contentWrapper div.league-container div.userInfo span.faintText")
    reputations=soup.select(".contentWrapper div.league-container div:nth-child(2) span.number")

    for rank,name,gold,silver,bronze,membership,reputation in zip(ranks,names,golds,silvers,bronzes,memberships,reputations):
        data.append({
            "Rank":rank.text,
            "Name":name.text,
            "Gold":gold.text,
            "Silver":silver.text,
            "Bronze":bronze.text,
            "Membership":membership.text,
            "Reputation":reputation.text
        })

    ndf=pd.DataFrame(data)
    df=pd.concat([df,ndf],ignore_index=True)
#df.head()
```

# Clean Data 

```python
df["Membership"]=(df.Membership.str.split(": ").str[1].str.split(",").str[0]+" and "+df.Membership.str.split(": ").str[1].str.split(",").str[1].str[:-5]).str.strip()
```

```python
df["Rank"]=df["Rank"].str.replace("#","").str.strip()
df["Reputation"]=df.Reputation.str.replace(",","")
df
```
