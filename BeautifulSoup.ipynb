# Bundesliga Champions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
import csv
```

```python
url="https://en.wikipedia.org/wiki/List_of_German_football_champions"
```

```python
page=requests.get(url).content
```

```python
soup=BeautifulSoup(page,"html.parser")
```

```python
seasons=[]
first_places=[]
runner_ups=[]
third_places=[]
top_scorrers=[]
```

```python
tbody=soup.find_all("tbody")[7]
trs=tbody.find_all("tr")
for tr in trs:
    season = tr.find_all("a")[0].text
    first_place = tr.find_all("a")[1].text
    runner_up = tr.find_all("a")[2].text
    third_place = tr.find_all("a")[3].text
    top_scorer = tr.find_all("a")[4].text
    seasons.append(season)
    first_places.append(first_place)
    runner_ups.append(runner_up)
    third_places.append(third_place)
    top_scorrers.append(top_scorer)      
```

```python
tuples=list(zip(seasons,first_places,runner_ups,third_places,top_scorrers))
```

```python
df=pd.DataFrame(tuples,columns=["Season","FirstPlace","RunnerUp","ThirdPlace","TopScorrer"])
```

# Seria A Table

```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
import mysql.connector

url = "https://www.skysports.com/serie-a-table"

page = requests.get(url).content
soup = BeautifulSoup(page, "html.parser")

tbody = soup.tbody
trs = tbody.find_all("tr")

with open("bundesliga.csv", "w", encoding="utf8", newline="") as f:
    thewriter = csv.writer(f)
    header = ["position", "team", "plays", "wins", "draws",
              "loses", "scored", "accepted", "goal_balance", "points"]
    thewriter.writerow(header)
    for tr in trs:
        positions = tr.find_all("td")[0]
        teams = tr.find_all("a", class_="standing-table__cell--name-link")
        for team in teams:
            plays = tr.find_all("td")[2]
            wins = tr.find_all("td")[3]
            draws = tr.find_all("td")[4]
            loses = tr.find_all("td")[5]
            scored = tr.find_all("td")[6]
            accepted = tr.find_all("td")[7]
            goal_balance = tr.find_all("td")[8]
            points = tr.find_all("td")[9]
        info = [positions.text, team.text, plays.text, wins.text, draws.text,
                loses.text, scored.text, accepted.text, goal_balance.text, points.text]
        thewriter.writerow(info)


df = pd.read_csv("bundesliga.csv")
df
```

# IMDB to MySQL Database

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import mysql.connector
import csv

url = "https://www.imdb.com/list/ls057336010/"

page = requests.get(url).content
soup = BeautifulSoup(page, "html.parser")

divs = soup.find_all("div", class_="lister-item-content")

with open("imdb100.csv", "w", encoding="utf8", newline="") as f:
    thewriter = csv.writer(f)
    header = ["Title", "Year", "Director"]
    thewriter.writerow(header)
    for div in divs:
        title = div.find("a")
        year = div.find("span", class_="lister-item-year text-muted unbold")
        runtime = div.find("span", class_="runtime")
        directors = div.find_all("p", class_="text-muted text-small")[1]
        name = directors.find("a")
        info = [title.text, year.text, name.text]
        thewriter.writerow(info)
df = pd.read_csv("imdb100.csv")
df

mydb = mysql.connector.connect(
    host="localhost", user="root", password="xxx", database="mydatabase")
mycursor = mydb.cursor()
mycursor.execute("Drop TABLE IF EXISTS table2")
mycursor.execute(
    "CREATE TABLE table2(title varchar(255), year varchar(55), director varchar(55))")
for row in df.itertuples():
    mycursor.execute("INSERT INTO table2(title, year, director) VALUES(%s,%s,%s)",
                     (row.Title, row.Year, row.Director))
mydb.commit()
```
