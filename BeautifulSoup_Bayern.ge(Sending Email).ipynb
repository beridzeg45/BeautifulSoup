# Let's scrape news data from bayrn.ge(official FC Bayern fan-club in Georgia) 

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

###  scrape news from bayern.ge(first 10 pages)

```python
news=pd.DataFrame()
for i in range(1,2):
    main_url="https://bayern.ge/siaxleebi/page/"+str(i)+"/"
    soup=bs(requests.get(main_url).content,"html.parser")
    links=soup.select("div[class='div-left-1'] div[class='news-7'] a")
    for link in links:
        url="https://bayern.ge/"+link["href"]
        soup=bs(requests.get(url).content,"html.parser")
        titles=soup.select("div[class='div-full-1'] div[class='div-full-4'] h1")
        articles=soup.select("div[class='div-full-1'] div[class='div-full-4'] p")
        times=soup.select("div[class='div-full-1'] div[class='div-full-9'] div")
        images=soup.select("div[class='div-full-1'] div[class='div-full-3'] a")
        data=[]
        for title,article,time,image in zip(titles,articles,times,images):
            data.append({
                "Title":title.text.strip(),
                "Article":article.text.strip(),
                "Published":time.text.strip(),
                "Link":url,
                "Image":image["href"]
            })
        df=pd.DataFrame(data)
        news=pd.concat([news,df],ignore_index=True)
```
